# GPT-2 Fine-tuning on Glaive Function Calling — Kaggle

Installs the [`gpt2fc` package](https://github.com/mron03/gpt2-function-calling) straight from GitHub and runs `gpt2fc-train`: downloads the OpenAI GPT-2 checkpoint, transfers the weights into the hand-written PyTorch model, and fine-tunes on [Glaive Function Calling v2](https://huggingface.co/datasets/glaiveai/glaive-function-calling-v2).

**Before running:**
- Accelerator → **GPU T4 x2**, Internet → **On**, then **Save Version → Save & Run All**
- Also works on Colab (**Runtime → Change runtime type → T4 GPU**)

Reference run: 355M, full 112,960-sample dataset, 1 epoch, batch size 8 → **~9.4 h** on a single T4. The defaults below are a small smoke test — flip `DATA_SLICE` to `-1` for the real thing.

```python
# TensorFlow (needed to read the original OpenAI checkpoints) is preinstalled
# on Kaggle and Colab, so the plain package is enough.
!pip install -q git+https://github.com/mron03/gpt2-function-calling

import torch
print('CUDA:', torch.cuda.is_available(), '—', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'enable the GPU accelerator!')
```

```python
MODEL_SIZE      = '124M'   # 124M / 355M / 774M / 1558M
NUM_EPOCHS      = 1
BATCH_SIZE      = 8
ALLOWED_MAX_LEN = 512
DATA_SLICE      = 100      # smoke test — set to -1 for the full dataset
OUTPUT_DIR      = 'checkpoints'
```

```python
import os

# Force a single GPU so the T4 x2 accelerator doesn't double-run anything
os.environ['CUDA_VISIBLE_DEVICES'] = '0'

!gpt2fc-train \
    --model-size {MODEL_SIZE} \
    --num-epochs {NUM_EPOCHS} \
    --batch-size {BATCH_SIZE} \
    --allowed-max-length {ALLOWED_MAX_LEN} \
    --data-slice {DATA_SLICE} \
    --output-dir {OUTPUT_DIR}
```

```python
ckpt = os.path.join(OUTPUT_DIR, f'gpt2-{MODEL_SIZE}-function-calling.pth')
print('Checkpoint:', ckpt)
print('Exists    :', os.path.exists(ckpt))
```

On Kaggle the checkpoint and loss plot land in `/kaggle/working/checkpoints` — download them from the **Output** tab. On Colab, grab them from the file browser (or mount Drive and point `OUTPUT_DIR` there before training).
